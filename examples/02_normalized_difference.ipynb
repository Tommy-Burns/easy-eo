# Computing Normalized Difference with Easy-EO

```python
from eeo import load_raster


red_path = r"./data/s2/L2A_B04_(Raw).tiff"
nir_path = r"./data/s2/L2A_B08_(Raw).tiff"


red = load_raster(red_path)
nir = load_raster(nir_path)
```

You can do the NDVI 2 ways `ds.normalized_difference(other)` or `normalized_difference(ds, other)`

### Method 1

```python
ndvi_1 = nir.normalized_difference(red)

ndvi_1.plot_raster(cmap="Greens", title="NDVI")
```

## Method 2

```python
from eeo.analysis import normalized_difference


ndvi_2 = normalized_difference(nir, red)
ndvi_2.plot_raster(title="NDVI", cmap="Greens")
```
