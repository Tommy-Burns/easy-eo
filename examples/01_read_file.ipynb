# Reading files with Easy-EO

## Reading a file from disk


```python
from eeo import load_raster


# Load raster from disk
path = "./data/dem.tiff"
ds = load_raster(path)
```

## Plot the raster

```python
ds.plot_raster(title="DEM", cmap="viridis")
```

## Plotting raster histogram

```python
ds.plot_histogram(title="Histogram")
```

## Plot raster with histogram

```python
ds.plot_raster_with_histogram(cmap="viridis")
```

## Close the datasets

```python
ds.close()
```
